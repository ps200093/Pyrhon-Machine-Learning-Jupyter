```python
import tensorflow as tf
import numpy as np

x_data = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]],dtype = np.float32)
y_data = np.array([[0],[1],[1],[0]],dtype = np.float32)

out_data = np.array([],dtype = np.float32)

W1 = tf.Variable(tf.random.normal([2, 2])) #hidden layer weight
b1 = tf.Variable(tf.random.normal([2]))


W2 = tf.Variable(tf.random.normal([2, 1]))#output layer weight
b2= tf.Variable(tf.random.normal([1]))

learning_rate = 0.1

def logisticRegression():
  with tf.GradientTape() as tape:
    layer1 = tf.sigmoid(tf.matmul(x_data, W1) + b1) # single perceptron
    layer2 = tf.sigmoid(tf.matmul(layer1, W2) + b2) # OUTPUT LAYER (MultiLayer perceptron)

    cost = -tf.reduce_mean(y_data * tf.math.log(layer2) + (1 - y_data) * tf.math.log(1 - layer2))

    gradients = tape.gradient(cost,(W2,b2,W1,b1))
    tf.optimizers.SGD(learning_rate).apply_gradients(zip(gradients,(W2,b2,W1,b1)))

for step in range(10001):
  logisticRegression()
  layer1 = tf.sigmoid(tf.matmul(x_data, W1) + b1)
  layer2 = tf.sigmoid(tf.matmul(layer1, W2) + b2)

  cost = -tf.reduce_mean(y_data * tf.math.log(layer2) + (1 - y_data) * tf.math.log(1 - layer2))

  prediction = tf.cast(layer2 > 0.5, dtype = tf.float32)

  accuaracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y_data), dtype = np.float32))

  if step % 2000 == 0:
    print("Step : ",step,
          "\n layer2 = ",layer2.numpy(),
          "\n b = ",b2.numpy(),
          "\n cost = ",cost.numpy(),
          "\n prediction = ",prediction.numpy(),
          "\n accuaracy = ", accuaracy.numpy(),
          "\n_________________________________")
```
